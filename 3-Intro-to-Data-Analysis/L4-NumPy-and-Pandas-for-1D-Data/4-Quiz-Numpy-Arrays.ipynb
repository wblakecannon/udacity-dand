```python
# This is my solution for the max_employment quiz
def max_employment(countries, employment):
    return (countries[employment.argmax()], employment[employment.argmax()])
max_employment(countries, employment)
```

```python
# This is the Udacity quiz solution and same code as a normal list
def max_employment2(countries, employment):
    max_country = None
    max_employment = 0
    
    for i in range(len(countries)):
        country = countries[i]
        country_employment = employment[i]
        if country_employment > max_employment:
            max_country = country
            max_employment = country_employment
    return (max_country, max_employment)
max_employment2(countries, employment)
```

```python
# This is Udacity's solution for the max_employment quiz
def max_employment3(countries, employment):
    i = employment.argmax()
    return (countries[i], employment[i])
max_employment3(countries, employment)
```
